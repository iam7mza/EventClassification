```python

import pandas as pd

# Load and analyze the data
df = pd.read_csv('/home/hamza/ML/mc.csv')

df
```

```python
print("Dataset shape:", df.shape)
print("\nColumn names:")
print(df.columns.tolist())
```

```python
# Find columns that contain list data (they should have brackets)
list_columns = []
for col in df.columns:
    if df[col].dtype == 'object':  # String columns that might contain lists
        sample_values = df[col].dropna().head(5).values
        if any('[' in str(val) for val in sample_values):
            list_columns.append(col)

print(f"\nColumns containing list data: {list_columns}")
```

```python
# Analyze each list column
for col in list_columns:
    print(f"\n--- Analysis of {col} ---")
    
    # Count empty lists
    empty_lists = df[df[col] == '[]'].shape[0]
    
    # Count single-element lists (approximate pattern)
    single_element = df[df[col].str.contains(r'^\[[^,]+\]$', na=False, regex=True)].shape[0]
    
    # Count multi-element lists
    multi_element = df[df[col].str.contains(r'.*,.*', na=False)].shape[0]
    
    print(f"Empty lists ([]): {empty_lists}")
    print(f"Single element lists: {single_element}")
    print(f"Multi-element lists: {multi_element}")
    print(f"Total non-null: {df[col].notna().sum()}")
    
    # Show some examples
    print(f"\nExamples of {col}:")
    print(df[col].dropna().head(3).values)
```
